```python
#pip install tensorflow 
import tensorflow as tf 
import numpy as np 
import matplotlib.pyplot as plt
from tensorflow.keras import models , layers
from tensorflow.keras.models import Model 
```

```python
print(tf.__version__)
```

```python
dataset=tf.keras.preprocessing.image_dataset_from_directory(
    "PlantVillage",
    shuffle=True,
    image_size=(256,256),
    batch_size=32,
)
```

```python
class_names=dataset.class_names
class_names
```

```python
for image_batch , label_batch in dataset.take(1):
    print("image_batch shape :",image_batch.shape)
    print("as wa can see there are three classes : class 0 ,class 1 and class 2 : ",label_batch.numpy())
    plt.figure(figsize=(14,14))
    for i in range(15):
        plt.subplot(3,5,i+1)
        plt.imshow(image_batch[i].numpy().astype("uint8"))
        plt.title(class_names[label_batch[i]])
```

```python
len(dataset) # 68*batch_size(=32)= 2176
```

```python
train = dataset.take(54)
test = dataset.skip(54).take(7)
validation = dataset.skip(54 + 7)
```

```python
input_shape=(256,256,3)
n_classes=3
model=models.Sequential([
    layers.Conv2D(32,(3,3),activation='relu',input_shape=input_shape),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(64,activation='relu'),
    layers.Dense(n_classes,activation='softmax'),
])

model.summary()
```

```python
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),optimizer='adam',metrics=['accuracy'])
```

```python
history= model.fit(train,
            verbose=1,
            validation_data=validation,
            batch_size=32,
            epochs=45)
```

```python
scores=model.evaluate(test)
```

```python
Epochs=45
acc=history.history['accuracy']
val_acc=history.history['val_accuracy']
loss=history.history['loss']
val_loss=history.history['val_loss']
```

```python
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.plot(range(Epochs),acc,label='Training accuracy')
plt.plot(range(Epochs),val_acc,label='validation accuracy')
plt.legend()
plt.title('Vizualization of the training and validation accuracy ')
plt.subplot(1,2,2)
plt.plot(range(Epochs),loss,label='Training loss')
plt.plot(range(Epochs),val_loss,label='validation loss')
plt.title('Vizualization of the training and validation loss ')
plt.legend()
```

```python
def prediction(model,img):
    img_array=tf.keras.preprocessing.image.img_to_array(img.numpy())
    #create a batch
    img_array=tf.expand_dims(img_array,0)
    predictions=model.predict(img_array)
    pred_class=class_names[np.argmax(predictions[0])]
    confidence=round(100*(np.max(predictions[0])),2)
    return pred_class , confidence
    
```

```python
plt.figure(figsize=(15,17))
for image_batch,label in test.take(1):
    for i in range(6) :
        plt.subplot(3,3,i+1)
        plt.imshow(image_batch[i].numpy().astype('uint8'))
        pred_class,confidence=prediction(model,image_batch[i])
        actual_class=class_names[label[i]]
        plt.title(f"Actual class :{actual_class},\n prediction :{pred_class},\n confidence :{confidence}")
```
