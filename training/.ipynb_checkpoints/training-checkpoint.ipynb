```python
import tensorflow as tf
from tensorflow.keras import models,layers
import matplotlib.pyplot as plt
import numpy as np 
print(tf.__version__)
```

```python
dataset=tf.keras.preprocessing.image_dataset_from_directory(
    "PlantVillage",
    shuffle=True,
    image_size=(256,256),
    batch_size=32,
)
```

```python
class_names=dataset.class_names
class_names
```

```python
for image_batch , label_batch in dataset.take(1):
    print("image_batch shape :",image_batch.shape)
    print("as wa can see there are three classes : class 0 ,class 1 and class 2 : ",label_batch.numpy())
    plt.figure(figsize=(14,14))
    for i in range(15):
        plt.subplot(3,5,i+1)
        plt.imshow(image_batch[i].numpy().astype("uint8"))
        plt.title(class_names[label_batch[i]])
```

#### train,test and validation dataset :



```python
len(dataset) # 68*batch_size(=32)= 2176
```

```python
# we want : 80% of dataset as training data 
# then the remaining dataset will be as : 10% for test and 10% for validation 
print("for training :",int(len(dataset)*0.8))
print("for test and validation :",len(dataset)-int(len(dataset)*0.8))
```

```python
train=dataset.take(54)
test=dataset.skip(54).take(7)
validation=dataset.skip(61)
print("train_size=",len(train),"\ntest_size=",len(test),"\nvalidation_size=",len(validation))
```

```python
trainds=train.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
test=test.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
validation=validation.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
def rescale(x, *args):
    return x / 255.0

# Perform manual rescaling using the helper function
train = trainds.map(rescale)
```

```python
resizing=tf.keras.Sequential([
    layers.experimental.preprocessing.Resizing(256,256),
])
```

```python
input_shape=(32,256,256,3)
model=models.Sequential([
    resizing,
    data_augmentation,
    layers.Conv2D(32,(3,3),activation='relu',input_shape=(32,256,256,3)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(64,activation='relu'),
    layers.Dense(3,activation='softmax')
])
```

```python
model.build(input_shape=input_shape)
```

```python
model.summary()
```

```python
# Data augmentation transforms
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.experimental.preprocessing.RandomFlip("horizontal_and_vertical"),
    tf.keras.layers.experimental.preprocessing.RandomRotation(0.2),
    tf.keras.layers.experimental.preprocessing.RandomZoom(0.2),
])
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(train,
         epochs=50,
         batch_size=32,
         verbose=1,
         validation_data=validation
         )
```
